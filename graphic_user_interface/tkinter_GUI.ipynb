```python
from tkinter import *
from tkinter import filedialog
from tensorflow.keras.preprocessing.image import img_to_array,array_to_img
import sys
import os
from PIL import ImageTk, Image

def run_():
    p01 = para01.get();p02 = para02.get();p03 = para03.get()
    p04 = para04.get();p05 = para05.get();p06 = para06.get()
    
    input_image = Image.open(image_data)
    numpy_image = img_to_array(input_image)
    check_shape=str(numpy_image.shape)
    
    output_text01 = "Paramters are "+p01+" , "+p02+" , "+p03+" , "\
    +p04+" , "+p05+" , "+p06+"\n"+" , and the image has shape of "+check_shape
    
    welcometext.configure(text="Please wait! GAN is running now..."+output_text01)
    
    ### GAN model generates a image in numpy array format ###
    ### get output image ###
    output_image = array_to_img(numpy_image)

    output_Window = Toplevel(root) 
    output_Window.title('Output image')
    output_Window.geometry('500x500')
    
    frame02 = Frame(output_Window , bg='white')
    frame02.place(relwidth=0.8, relheight=0.8, relx=0.1, rely=0.1)

    for img_display in frame02.winfo_children():
        img_display.destroy()
        
    output_image = Image.open(image_data) # image_data is a file object
    output_image = output_image.resize((300, 500), Image.ANTIALIAS)
    output_image = ImageTk.PhotoImage(output_image)
    
    panel_image02 = Label(frame02, image=output_image).pack() # show image on the frame
    panel_image02.pack(side = "bottom", fill = "both", expand = "yes")
    
def restart_():
    """Restarts the current program.
    Note: this function does not return. Any cleanup action (like
    saving data) must be done before calling this function."""
    python = sys.executable
    os.execl(python, python, * sys.argv)
    

def reset_():
    global para01, para02, para03, para04, para05, para06
    para04.delete(0, END); para05.delete(0, END); para06.delete(0, END);
    para04.insert(0,0); para05.insert(0,0); para06.insert(0,0) 
    
    para01.delete(0, END); para02.delete(0, END); para03.delete(0, END);
    para01.insert(0,1); para02.insert(0,1); para03.insert(0,1)
    
def help_():
    help_Window = Toplevel(root) 
    help_Window.title('Help')
    help_Window.geometry('400x400')

    helptext = Label(help_Window, text="If you need help, please visit this website")
    helptext.place(x=5, y=5)
    
def load_img():
    global img, image_data
    for img_display in frame.winfo_children():
        img_display.destroy()

    image_data = filedialog.askopenfilename(initialdir="/", title="Choose an image",
                                       filetypes=(("all files", "*.*"), ("png files", "*.png")))
    basewidth = 250 # Processing image for dysplaying
    img = Image.open(image_data)
    wpercent = (basewidth / float(img.size[0]))
    hsize = int((float(img.size[1]) * float(wpercent)))
    img = img.resize((basewidth, hsize), Image.ANTIALIAS)
    img = ImageTk.PhotoImage(img)
    file_name = image_data.split('/')
    panel = Label(frame, text= str(file_name[len(file_name)-1]).upper()).pack()
    panel_image = Label(frame, image=img).pack()
    
    
root = Tk()
root.title('Generative adversarial network')
root.geometry('500x500')

welcometext = Label(root, text="Hello! Welcome to our GAN model!")
welcometext.place(x=50, y=5)

# add run button
btn_run = Button(root, text="Run", padx=10,pady=0,width=3, bg="green", fg="black", command=run_)
btn_run.place(x=160,y=470)

# add reset button
btn_reset = Button(root, text="Reset", padx=10,pady=0,width=3, bg="green", fg="black", command=reset_)
btn_reset.place(x=240,y=470)

# add quit button
btn_quit = Button(root, text="Quit", padx=10,pady=0,width=3, bg="green", fg="black", command=quit)
btn_quit.place(x=320,y=470)

# add help button
btn_help = Button(root, text="Help", padx=10,pady=0,width=3, bg="green", fg="black", command=help_)
btn_help.place(x=400,y=470)


# add parameter 01
para01_lbl = Label(root, text="parameter01")
para01_lbl.place(x=10,y=300)
para01 = Spinbox(root, from_=1, to=50, width=10)
para01.place(x=10,y=320)

# add parameter 02
para02_lbl = Label(root, text="parameter02")
para02_lbl.place(x=180,y=300)
para02 = Spinbox(root, from_=1, to=50, width=10)
para02.place(x=180,y=320)

# add parameter 03
para03_lbl = Label(root, text="parameter03")
para03_lbl.place(x=350,y=300)
para03 = Spinbox(root, from_=1, to=50, width=10)
para03.place(x=350,y=320)

# add parameter 04
para04_lbl = Label(root, text="parameter04")
para04_lbl.place(x=10,y=360)
para04 = Entry(root,width=10)
para04.place(x=10,y=380)
para04.insert(0,0) # set the default value as 0

# add parameter 05
para05_lbl = Label(root, text="parameter05")
para05_lbl.place(x=180,y=360)
para05 = Entry(root,width=10)
para05.place(x=180,y=380)
para05.insert(0,0) # set the default value as 0

# add parameter 06
para06_lbl = Label(root, text="parameter06")
para06_lbl.place(x=350,y=360)
para06 = Entry(root,width=10)
para06.place(x=350,y=380)
para06.insert(0,0) # set the default value as 0


# add Load_image button
btn_loadimage = Button(root, text="Load image",padx=10,pady=0,width=6, \
                       bg="green", fg="black", command=load_img)
btn_loadimage.place(x=10,y=48)

frame = Frame(root, bg='white')
frame.place(relwidth=0.6, relheight=0.5, relx=0.3, rely=0.1)

root.mainloop()
```
